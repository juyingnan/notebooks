```python
def simple_stats(item_list):
    item_cat = list(set(item_list))
    item_count = []
    for i in range(len(item_cat)):
        item_count.append(0)
        _item = item_cat[i]
        for item in item_list:
            if item == _item:
                item_count[i] += 1
    return item_cat, item_count
```

```python
import re
normalized_column_software = []
for s_list in column_software:
    normalized_column_software.append([])
    for item in re.split(';| and |,|\n', s_list):
        if len(item.strip().lower().title()) > 0:
            for k, v in software_dict.items():
                if re.match(k,item.lower()):
                    normalized_column_software[-1].append(v)
                    break
            else:
                normalized_column_software[-1].append(item.strip('.').strip().lower().title())
for i in range(len(normalized_column_software)):
    normalized_column_software[i] = sorted(list(set(normalized_column_software[i])))
# normalized_column_software
```

```python
from string import punctuation

software_cat, software_count = simple_stats([item for sub_list in normalized_column_software for item in sub_list ])

print(software_cat)        
print(software_count)
```

```python
def stacked(df):
    df_top = df.cumsum(axis=1)
    df_bottom = df_top.shift(axis=1).fillna({list(df)[0]: 0})[::-1]
    df_stack = pd.concat([df_bottom, df_top], ignore_index=True)
    return df_stack
```

```python
import networkx as nx
G_list = list()

for year in year_list:

    G = nx.Graph()

    for student in yearly_software_column[year]:
        for i in range(len(student)):
            software = student[i]
            if G.has_node(software):
                nx.set_node_attributes(G, {software:{'size':G.nodes[software]['size']+1}})
            else:
                G.add_node(software, size = 2, label = software)
    for student in yearly_software_column[year]:
        for i in range(len(student)):
            software = student[i]
            if i + 1 < len(student):
                for j in range(i + 1, len(student)):
                    if G.has_edge(software, student[j]):
                        nx.set_edge_attributes(G, {(software, student[j]): 
                                                   {'weight':G.edges[software, student[j]]['weight']+1}})
                    else:
                        G.add_edge(software, student[j], weight = 1)
    # edge filter
    remove_edges = [(start, end) for start, end, data in G.edges.data() if data['weight'] <= 1]
    G.remove_edges_from(remove_edges)
    
    #node filter
    remove_nodes = list(nx.isolates(G))
    G.remove_nodes_from(remove_nodes)
    
    G_list.append(G)
```

```python
from bokeh.models.graphs import from_networkx
from bokeh.models import LabelSet, Circle, MultiLine
from bokeh.transform import linear_cmap
from bokeh import palettes
from networkx.drawing.nx_agraph import graphviz_layout
import random
output_notebook()

tab_list = list()
for G, year in zip(G_list, year_list):
    graph = from_networkx(G, graphviz_layout, prog='neato')
    graph.node_renderer.data_source.data['node_size'] = [item * 1 for item in graph.node_renderer.data_source.data['size']]
    # graph.node_renderer.data_source.data['node_color'] = palettes.viridis(len(G.nodes)*2)[len(G.nodes):]
    degrees = [item[1] for item in G.degree]
    graph.node_renderer.data_source.data['degree'] = degrees
    # graph.node_renderer.data_source.data['node_color'] = [random.choice(palettes.Pastel1[9]) for i in range(len(G.nodes))]
    # graph.node_renderer.data_source.data['node_color'] = [palettes.magma(len(G.nodes)*2))[] for i in range(len(G.nodes))]
    mapper = linear_cmap(field_name='degree', palette=palettes.Plasma[11] ,low=min(degrees) ,high=max(degrees))

    labels = nx.get_node_attributes(G, 'label')
    sizes = nx.get_node_attributes(G, 'size')
    x, y = zip(*graph.layout_provider.graph_layout.values())
    data = {'x': list(x), 'y': list(y), 'label': [], 'size': []}
    for node in labels.keys():
        data['label'].append(labels[node])
        data['size'].append(sizes[node])
    source = ColumnDataSource(data)
    labels = LabelSet(x='x', y='y', text='label', source=source, level='glyph',render_mode='canvas',
                      x_offset=-15, y_offset=15)

    plot = figure(title="Software used", x_range=(min(x)-20,max(x)+20), y_range=(min(y)-20,max(y)+20),
                  tools=tools_list, 
                  tooltips= [("Software", "@label"),
                             ("Used", "@size"),
                             ("Link", "@degree"),])
    plot.renderers.append(graph)
    graph.node_renderer.glyph.size = {'field': 'node_size'}
    graph.node_renderer.glyph.fill_color = mapper#{'field': 'node_color'}
    # graph.node_renderer.glyph.fill_alpha = 0.5
    graph.edge_renderer.glyph.line_width = {'field': 'weight'}
    graph.edge_renderer.glyph.line_alpha = 0.3

    plot.renderers.append(labels)
    tab_list.append(Panel(child=plot, title=year))

tabs = Tabs(tabs=tab_list)
show(tabs)
```
